```python
import pandas as pd
train = pd.read_csv("http://archive.ics.uci.edu/ml/machine-learning-databases/madelon/MADELON/madelon_train.data", delimiter=' ', header=None)
train_scaled = pd.read_csv("../datasets/new_madelon_train.csv")
y_train = pd.read_csv("http://archive.ics.uci.edu/ml/machine-learning-databases/madelon/MADELON/madelon_train.labels", delimiter=' ', header=None)
```

```python
train_scaled = train_scaled.drop(columns=["Unnamed: 0"])
y_train = y_train[0].ravel()
train_scaled.head()
```

```python
import seaborn as sns
import numpy as np
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.metrics import recall_score, make_scorer
from sklearn.model_selection import cross_val_score, cross_val_predict

# loop to find best number of estimators
min_estimators = 100
max_estimators = 200
step = 25
result = []
best_score = 0
best_estimator = 0

for i in range(min_estimators, max_estimators+step, step):
    clf = GradientBoostingClassifier(
        n_estimators=i, 
        random_state=0)
    scores = cross_val_score(clf, train_scaled, y_train, cv=3)
    if (scores.mean() > best_score):
        best_estimator = i
        best_score = scores.mean()
    result.append((i, scores.mean()))
    
# converting result into dataframe
estimators = np.array(result)[:,0]
score = np.array(result)[:,1]
d = {'estimators': estimators, 'score': score}
df_scores = pd.DataFrame(d)
print(f'Best estimator: {best_estimator}')
print(df_scores)

# plotting results
sns.set_theme(style="dark")
sns.set_palette("colorblind")
sns.lineplot(
        data=df_scores,
        x="estimators", 
        y="score"
    )
```

```python
clf = GradientBoostingClassifier(
        n_estimators=100,
        random_state=0)
scores = cross_val_score(clf, train_scaled, y_train, cv=3)
y_pred = cross_val_predict(clf, train_scaled, y_train, cv=3)
print("recall_score: %0.5f (+/- %0.5f)" % (scores.mean(), scores.std()))
```

```python
from sklearn.metrics import confusion_matrix
from sklearn.metrics import classification_report, plot_confusion_matrix

print(confusion_matrix(y_train, y_pred))
print(classification_report(y_train, y_pred))
```
